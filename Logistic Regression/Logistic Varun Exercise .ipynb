```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Bankruptcy_data.csv')
df
```

```python
df.shape
```

```python
from sklearn import preprocessing
```

```python
from sklearn .preprocessing import Imputer
```

```python
imputer= Imputer()
df1= pd.DataFrame (imputer.fit_transform(df))
df1.isnull().sum()
```

```python
x=df1.iloc[:,:-1]
x
```

```python
y=df1.iloc[:,-1:]
y
```

```python
from sklearn.linear_model import LogisticRegression
Logmodel= LogisticRegression()
```

```python
Log_model=Logmodel.fit(x,y)
```

```python
Log_model.coef_
```

```python
Log_model.intercept_
```

```python
y_pred= Log_model.predict(x)
y_pred
```

```python
Log_model.score(x,y)
```

```python
from sklearn.metrics import classification_report 
```

```python
print (classification_report (y, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix=confusion_matrix(y,y_pred)
print(confusion_matrix)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
Logit_roc_score=roc_auc_score(y,Log_model.predict(x))
Logit_roc_score
```

```python
fpr, tpr, thresholds = roc_curve(y,Log_model.predict_proba(x)[:,-1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area=%0.20f)'% Logit_roc_score)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Rectified operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()


        
         

```
